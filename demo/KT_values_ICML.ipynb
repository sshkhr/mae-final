```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dfs = {'All':[], 'Correct':[], 'Incorrect':[]}
```

```python
xls = pd.ExcelFile('KT_values.xlsx')
dfs['All'] = pd.read_excel(xls, 'Sheet15')
dfs['All'].rename( columns={'Unnamed: 0':'Models'}, inplace=True )
dfs['Correct'] = pd.read_excel(xls, 'Sheet16')
dfs['Correct'].rename( columns={'Unnamed: 0':'Models'}, inplace=True )
dfs['Incorrect'] = pd.read_excel(xls, 'Sheet17')
dfs['Incorrect'].rename( columns={'Unnamed: 0':'Models'}, inplace=True )
```

```python
dfs['All']
```

```python
fig_dir = '../figures/ICML/'
```

```python
fig, axn = plt.subplots(1, 3, figsize=(28, 11), dpi=500, sharex=True, sharey=False)
#cbar_ax = fig.add_axes([.9, .3, .01, .5])

#fig.supxlabel('common_x')
fig.supylabel('Kendall\'s Tau Rank Correlation', fontweight="bold", fontsize = 30, x=0.07)

sns.set(font_scale=1.5)

index = {0:'All', 1:'Correct', 2:'Incorrect'}
for i, ax in enumerate(axn.flat):
    print(index[i])
    df = dfs[index[i]].melt(id_vars=["Models"], value_vars=['Top-5 ranks', 'Top-10 ranks'], var_name="Ranking", value_name="KT Rank Correlation")

    plot = sns.barplot(data=df, x='Ranking', y='KT Rank Correlation', hue='Models', ax=ax)

    ax.set_ylabel('')
    ax.set_xlabel('')

    ax.tick_params(axis='x', which='major', labelsize=30)
    ax.tick_params(axis='y', which='major', labelsize=25)
    
    if(i<2):
        ax.get_legend().remove()
    else:
        leg = ax.legend(fontsize=30)
        leg.set_title('Models',prop={'size':30, 'weight':'bold'})


    letter = {0:'(A) ', 1:'(B) ', 2:'(C) '}
    ax.set_title(letter[i] +index[i]+" predictions", fontdict={'fontweight':"bold", 'fontsize':30}, y=-0.2)

plt.show()

fig.savefig(fig_dir + 'KT_plot.pdf', dpi=500, bbox_inches = "tight")
```
